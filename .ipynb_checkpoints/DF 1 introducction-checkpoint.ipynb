# 7 
## Drop DataFrame Rows with Null Values with the dropna Method

```python
nba = pd.read_csv('nba.csv')
```

```python
nba.tail()
```

```python
nba.dropna()
```

```python
nba.dropna(how = 'any') # si en la fila hay algún nulo
```

```python
nba.dropna(how = 'all') # si toda la fila es nula
```

```python
nba.dropna(subset = ['College'])
```

# 8
## Fill in Missing DataFrame Values with the fillna Method

```python
nba = pd.read_csv('nba.csv')
```

```python
nba['College'].fillna('Unknown',inplace = True)
```

```python
nba.tail()
```

```python
nba['Salary'].fillna(value = 0, inplace = True)
```

```python
nba.tail()
```

```python
nba['Salary'] = nba['Salary'].fillna(value = 0)
```

```python
nba.tail()
```

# 9 

## The astype Method I

```python
nba = pd.read_csv('nba.csv').dropna(how = 'all')
```

```python
nba['Age'].hasnans
```

```python
nba['Age'].astype("int")
```

```python
nba['Age'] = nba['Age'].astype("int")
nba.head()
```

```python
nba['Salary'].hasnans
```

```python
nba['Salary'] = nba['Salary'].fillna(0) # sino rellenamos el campo con nan antes de hacer la trasnformación nos dará un error.
```

```python
nba['Salary'] = nba['Salary'].astype(int)
```

```python
nba.head()
```

# 13 
### Rank Series Values with the rank Method

```python
nba['Salary'] = nba['Salary'].fillna(value = 0)
nba['Salary'] = nba['Salary'].astype(int)
```

```python
nba['Salary'].rank(ascending = True)
```

```python
nba['Salary Rank'] = nba['Salary'].rank(ascending = True).astype("int")
```

```python
nba.sort_values("Salary",ascending = False)
```
