# 4
### Extract Rows from a MultiIndex DataFrame

```python
df = pd.read_csv('./data/bigmac.csv',parse_dates = ['Date'], index_col = ["Date", 'Country'])
df.sort_index(inplace = True)
df.head()
```

```python
df.loc['2010-01-01']
```

```python
df.loc['2010-01-01','Argentina']
```

```python
df.loc[('2010-01-01','Argentina'),'Price in US Dollars']
```

```python
df.loc[('2010-01-01','Argentina'),['Price in US Dollars','Price in US Dollars']]
```

```python
df.loc[('2010-01-01',)]
```

# 6
### The swaplevel Method


```python
df = pd.read_csv('./data/bigmac.csv',parse_dates = ['Date'], index_col = ["Date", 'Country'])
df.sort_index(inplace = True)
df.head()
```

```python
df.swaplevel()
```

```python
df.swaplevel("Country","Date")
```

# 7 
### The stack() Method

```python
df = pd.read_csv("./data/worldstats.csv",index_col = ["country","year"])
df.head()
```

```python
df.stack()
```

```python
df.stack().to_frame()
```

```python
s = df.stack()
s.head()
```

```python
s.unstack().unstack().unstack()
```

```python
s.head()
```

```python
s.unstack(2)
```

# 8
### Pivot

```python
df = pd.read_csv('./data/salesmen.csv')
df['Salesman'] = df['Salesman'].astype("category")
df.head()
```

```python
df['Salesman'].value_counts()
```

```python
df.pivot(index = "Date",columns = 'Salesman')
```

# 9
### Pivot_table()

```python
df = pd.read_csv('./data/foods.csv')
df.head()
```

```python
df.pivot_table(values = 'Spend',index = 'Gender',aggfunc = 'mean')
```

```python
df.pivot_table(values = 'Spend',index = 'Item',aggfunc = 'sum')
```
